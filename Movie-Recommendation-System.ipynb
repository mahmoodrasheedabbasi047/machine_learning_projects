```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

movies.head
```

```python
credits.head(1)
```

```python
data = movies.merge(credits, on='title')
data.shape
```

```python
data.info()
```

```python
data = data[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
data.sample(5)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.iloc[0].genres
```

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ["Action", "Adventure", "Fantaasy", "SciFi"]

import ast

def convert(obj):
    gen = []
    
    for i in ast.literal_eval(obj):
        gen.append(i['name'])
    
    return gen

# convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
data.genres = data.genres.apply(convert)
data
```

```python
data.keywords = data.keywords.apply(convert)
```

```python
data
```

```python
def getActors(obj):
    actors = []
    count_actors = 0
    
    for i in ast.literal_eval(obj):
        if count_actors !=3:
            actors.append(i['name'])
        else:
            break
    
    return actors

data.cast = data.cast.apply(getActors)
```

```python
data
```

```python
data.iloc[0].crew
```

```python
def getDirector(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l

data.crew = data.crew.apply(getDirector)
```

```python
data
```

```python
data.overview = data.overview.apply(lambda x : x.split())
```

```python
data
```

```python
# Removing Spaces

data.genres.apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
data.genres = data.genres.apply(lambda x:[i.replace(" ", "") for i in x])
data.keywords = data.keywords.apply(lambda x:[i.replace(" ", "") for i in x])
data.overview = data.overview.apply(lambda x:[i.replace(" ", "") for i in x])
data.crew = data.crew.apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
data.head()
```

```python
# corpus = collection of columns
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
data
```

```python
movies_data = data[['movie_id', 'title', 'tags']]
movies_data
```

```python
movies_data['tags']
```

```python
movies_data['tags'] = movies_data['tags'].apply(lambda x : " ".join(x))
movies_data
```

```python
# lower case conversion
movies_data['tags'] = movies_data['tags'].apply(lambda x:x.lower())
movies_data
```

```python
# Text vectorization
# vector contains no:of occourance of a word
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies_data['tags']).toarray()
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
movies_data['tags'] = movies_data['tags'].apply(stem)
movies_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies_data['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
# Getting similarities one movies to others

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity
```

```python
movies_data[movies_data['title'] == 'Avatar']
```

```python
# It is not valid to get way top 5 recommended movies, because index are removed so how to access
sorted(similarity[0], reverse=True)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recomend(movie):
    movie_index = movies_data[movies_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies_data.iloc[i[0]].title)
    
```

```python
recomend('Batman Begins')
```
